## Preprocessing

### Subtask:
Prepare the text data for BERT by handling missing values, tokenization, and creating appropriate input formats (input IDs, attention masks, etc.).


**Reasoning**:
Identify the relevant text columns, handle missing values, define a simple target variable for demonstration, combine the text data, initialize a BERT tokenizer, tokenize the text data, and create PyTorch tensors.

```python
import torch
from transformers import BertTokenizer

# 1. Identify relevant text columns (Corrected based on df.columns output and character fix)
text_cols = [
    '1. How would you describe your overall mental health at this moment?  ',
    '2. What factors do you think contribute the most to students/working professionals experiencing mental health issues?  ',
    "3. In your opinion, how does academic/work-life pressure affect students' or individual's mental health?  ",
    '4. How comfortable do you feel discussing mental health and depression with your peers or teachers or colleague? Why or why not? ',
    "5. Suppose your parents or guardians don’t understand your mental health struggles and think you're just being “lazy” or “dramatic.” How would you handle that situation emotionally and practically?  ",
    '6. How do you cope with stress or difficult emotions during challenging times? ',
    '7.  Do you often find yourself expecting negative outcomes in life ? Please elaborate.',
    '8. How does your personal relationships (family, friends, romantic) influence your mental well-being? ',
    '9. Have strained relationships ever contributed to feelings of sadness, loneliness, or depression for you? Please explain',
    '10. If you notice yourself thinking negatively about the future or feeling hopeless for several days, how would you interpret these signs? Would you seek help?  '
]

# 2. Handle missing values in the selected text columns
df[text_cols] = df[text_cols].fillna('')

# 3. Define a target variable for depression detection (simple approach for demonstration)
# Label responses containing keywords related to negative feelings or seeking help as potentially depressed (1), otherwise 0.
negative_keywords = ['bad', 'stress', 'pressure', 'negative', 'sadness', 'loneliness', 'depression', 'hopeless', 'not good', 'messed up', 'need help', 'seek help']

def label_depression(row):
    combined_text = ' '.join(str(row[col]).lower() for col in text_cols)
    for keyword in negative_keywords:
        if keyword in combined_text:
            return 1
    return 0

df['potential_depression'] = df.apply(label_depression, axis=1)

# 4. Combine text from selected columns
df['combined_text'] = df[text_cols].agg(' '.join, axis=1)

# 5. Initialize a BERT tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# 6. Tokenize the combined text data
max_seq_length = 128 # Choose a suitable sequence length

encoded_inputs = tokenizer(
    list(df['combined_text']),
    add_special_tokens=True,
    padding='max_length',
    truncation=True,
    max_length=max_seq_length,
    return_attention_mask=True,
    return_tensors='pt' # Return PyTorch tensors
)

input_ids = encoded_inputs['input_ids']
attention_masks = encoded_inputs['attention_mask']

# 7. Create PyTorch tensors for the target variable
labels = torch.tensor(df['potential_depression'].values)

print("Shape of input_ids:", input_ids.shape)
print("Shape of attention_masks:", attention_masks.shape)
print("Shape of labels:", labels.shape)
```

## Training

### Subtask:
Train the BERT model on the preprocessed data.


**Reasoning**:
Split the data, create DataLoaders, define optimizer and scheduler, train the model, evaluate, and save the model.



```python
from torch.utils.data import TensorDataset, DataLoader, RandomSampler, SequentialSampler
from transformers import get_linear_schedule_with_warmup
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_recall_fscore_support
import numpy as np
import random
import time
import datetime
from torch.optim import AdamW
import os

# Helper function for formatting time
def format_time(elapsed):
    elapsed_rounded = int(round((elapsed)))
    return str(datetime.timedelta(seconds=elapsed_rounded))

# 1. Split the data into training and validation sets
train_inputs, validation_inputs, train_labels, validation_labels, train_masks, validation_masks = train_test_split(
    input_ids, labels, attention_masks, random_state=2018, test_size=0.1
)

# 2. Create PyTorch DataLoader objects
batch_size = 32

train_data = TensorDataset(train_inputs, train_masks, train_labels)
train_sampler = RandomSampler(train_data)
train_dataloader = DataLoader(train_data, sampler=train_sampler, batch_size=batch_size)

validation_data = TensorDataset(validation_inputs, validation_masks, validation_labels)
validation_sampler = SequentialSampler(validation_data)
validation_dataloader = DataLoader(validation_data, sampler=validation_sampler, batch_size=batch_size)

# 3. Define an optimizer and a learning rate scheduler
optimizer = AdamW(model.parameters(), lr=5e-5, eps=1e-8)

epochs = 4 # Reduced epochs for faster execution
total_steps = len(train_dataloader) * epochs
scheduler = get_linear_schedule_with_warmup(optimizer, num_warmup_steps=0, num_training_steps=total_steps)

# Function to calculate accuracy
def flat_accuracy(preds, labels):
    pred_flat = np.argmax(preds, axis=1).flatten()
    labels_flat = labels.flatten()
    return np.sum(pred_flat == labels_flat) / len(labels_flat)

# Set the seed value for reproducibility.
seed_val = 42
random.seed(seed_val)
np.random.seed(seed_val)
torch.manual_seed(seed_val)
torch.cuda.manual_seed_all(seed_val)

# Store the average loss after each epoch so we can plot them.
loss_values = []

# 4. Train the model
for epoch_i in range(0, epochs):
    print(f'======== Epoch {epoch_i + 1} / {epochs} ========')
    print('Training...')

    t0 = time.time()
    total_loss = 0
    model.train() # Put the model into training mode

    for step, batch in enumerate(train_dataloader):
        # Progress update every 40 batches.
        if step % 40 == 0 and not step == 0:
            elapsed = format_time(time.time() - t0)
            print(f'  Batch {step} of {len(train_dataloader)}. Elapsed: {elapsed}.')

        b_input_ids = batch[0].to(model.device)
        b_input_mask = batch[1].to(model.device)
        b_labels = batch[2].to(model.device)

        model.zero_grad() # Clear previously calculated gradients

        outputs = model(b_input_ids,
                        token_type_ids=None,
                        attention_mask=b_input_mask,
                        labels=b_labels)

        loss = outputs[0]
        total_loss += loss.item()

        loss.backward() # Perform a backward pass to calculate the gradients.
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0) # Clip the norm of the gradients to 1.0 to prevent the "exploding gradients" problem.
        optimizer.step() # Update parameters
        scheduler.step() # Update the learning rate.

    avg_train_loss = total_loss / len(train_dataloader)
    loss_values.append(avg_train_loss)

    print(f"  Average training loss: {avg_train_loss:.2f}")
    print(f"  Training epoch took: {format_time(time.time() - t0)}")

    # 5. Evaluation on the validation set
    print("Running Validation...")
    t0 = time.time()
    model.eval() # Put the model in evaluation mode

    eval_accuracy = 0
    nb_eval_steps = 0
    all_preds = []
    all_labels = []

    for batch in validation_dataloader:
        batch = tuple(t.to(model.device) for t in batch)
        b_input_ids, b_input_mask, b_labels = batch

        with torch.no_grad(): # Don't compute gradients
            outputs = model(b_input_ids,
                            token_type_ids=None,
                            attention_mask=b_input_mask)

        logits = outputs[0]
        logits = logits.detach().cpu().numpy()
        label_ids = b_labels.to('cpu').numpy()

        eval_accuracy += flat_accuracy(logits, label_ids)
        nb_eval_steps += 1

        preds = np.argmax(logits, axis=1).flatten()
        all_preds.extend(preds)
        all_labels.extend(label_ids)

    avg_val_accuracy = eval_accuracy / nb_eval_steps
    print(f"  Accuracy: {avg_val_accuracy:.2f}")
    print(f"  Validation took: {format_time(time.time() - t0)}")

    # Calculate Precision, Recall, F1-score
    precision, recall, f1, _ = precision_recall_fscore_support(all_labels, all_preds, average='binary')
    print(f"  Precision: {precision:.2f}")
    print(f"  Recall: {recall:.2f}")
    print(f"  F1-score: {f1:.2f}")


print("")
print("Training complete!")

# 6. Save the trained model
output_dir = './model_save/'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

print(f"Saving model to {output_dir}")
model_to_save = model.module if hasattr(model, 'module') else model
model_to_save.save_pretrained(output_dir)
tokenizer.save_pretrained(output_dir)
```

## Evaluation

### Subtask:
Evaluate the performance of the trained model using appropriate metrics.


**Reasoning**:
Evaluate the performance of the trained model using the validation data and calculate evaluation metrics.



```python
# The model and tokenizer are already loaded and available from the previous step.
# The validation data (validation_inputs, validation_masks, validation_labels)
# and the validation_dataloader are also available from the previous step.

print("Running Final Evaluation...")
t0 = time.time()

# Put the model in evaluation mode
model.eval()

# Initialize variables to store evaluation metrics
eval_accuracy = 0
nb_eval_steps = 0
all_preds = []
all_labels = []

# Iterate through the validation data
for batch in validation_dataloader:
    # Move batch to the appropriate device
    batch = tuple(t.to(model.device) for t in batch)
    b_input_ids, b_input_mask, b_labels = batch

    # Perform forward pass
    with torch.no_grad():
        outputs = model(b_input_ids,
                        token_type_ids=None,
                        attention_mask=b_input_mask)

    # Get logits and move to CPU
    logits = outputs[0]
    logits = logits.detach().cpu().numpy()
    label_ids = b_labels.to('cpu').numpy()

    # Calculate accuracy
    eval_accuracy += flat_accuracy(logits, label_ids)
    nb_eval_steps += 1

    # Get predictions and store
    preds = np.argmax(logits, axis=1).flatten()
    all_preds.extend(preds)
    all_labels.extend(label_ids)

# Calculate average validation accuracy
avg_val_accuracy = eval_accuracy / nb_eval_steps
print(f"  Final Accuracy: {avg_val_accuracy:.2f}")

# Calculate and print Precision, Recall, and F1-score
precision, recall, f1, _ = precision_recall_fscore_support(all_labels, all_preds, average='binary')
print(f"  Final Precision: {precision:.2f}")
print(f"  Final Recall: {recall:.2f}")
print(f"  Final F1-score: {f1:.2f}")

print(f"Evaluation took: {format_time(time.time() - t0)}")
```

```python
import torch
from transformers import BertForSequenceClassification, get_linear_schedule_with_warmup
from torch.optim import AdamW
import numpy as np
import random
import time
import datetime
import os
from sklearn.metrics import accuracy_score, precision_recall_fscore_support

# Helper function for formatting time (assuming it was defined earlier, re-defining for self-containment if not)
def format_time(elapsed):
    elapsed_rounded = int(round((elapsed)))
    return str(datetime.timedelta(seconds=elapsed_rounded))

# Function to calculate accuracy (assuming it was defined earlier, re-defining for self-containment if not)
def flat_accuracy(preds, labels):
    pred_flat = np.argmax(preds, axis=1).flatten()
    labels_flat = labels.flatten()
    return np.sum(pred_flat == labels_flat) / len(labels_flat)

# 1. Re-initialize a BertForSequenceClassification model
num_labels = 2 # Binary classification (0 or 1 for depression)
model_new = BertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=num_labels)

# Move the model to GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model_new.to(device)
print(f"New model initialized and moved to {device}.")

# 2. Define an AdamW optimizer for the new model
optimizer_new = AdamW(model_new.parameters(), lr=5e-5, eps=1e-8)

# 3. Set the number of training epochs and calculate total training steps
epochs = 4 # As suggested, reduced epochs for faster execution
total_steps_new = len(train_dataloader_new) * epochs

# 4. Initialize a learning rate scheduler
scheduler_new = get_linear_schedule_with_warmup(optimizer_new, num_warmup_steps=0, num_training_steps=total_steps_new)

# Set the seed value for reproducibility (important for consistent results)
seed_val = 42
random.seed(seed_val)
np.random.seed(seed_val)
torch.manual_seed(seed_val)
torch.cuda.manual_seed_all(seed_val)

# Store the average loss after each epoch
loss_values_new = []

# 5. Implement the training loop
print("\nStarting training for the new model...")
for epoch_i in range(0, epochs):
    print(f'======== Epoch {epoch_i + 1} / {epochs} ========')
    print('Training...')

    t0 = time.time()
    total_loss_new = 0
    model_new.train() # Put the model into training mode

    for step, batch in enumerate(train_dataloader_new):
        if step % 40 == 0 and not step == 0:
            elapsed = format_time(time.time() - t0)
            print(f'  Batch {step} of {len(train_dataloader_new)}. Elapsed: {elapsed}.')

        b_input_ids = batch[0].to(device)
        b_input_mask = batch[1].to(device)
        b_labels = batch[2].to(device)

        model_new.zero_grad() # Clear previously calculated gradients

        outputs = model_new(b_input_ids,
                              token_type_ids=None,
                              attention_mask=b_input_mask,
                              labels=b_labels)

        loss = outputs[0]
        total_loss_new += loss.item()

        loss.backward() # Perform a backward pass to calculate the gradients.
        torch.nn.utils.clip_grad_norm_(model_new.parameters(), 1.0) # Clip gradients
        optimizer_new.step() # Update parameters
        scheduler_new.step() # Update the learning rate.

    avg_train_loss_new = total_loss_new / len(train_dataloader_new)
    loss_values_new.append(avg_train_loss_new)

    print(f"  Average training loss: {avg_train_loss_new:.2f}")
    print(f"  Training epoch took: {format_time(time.time() - t0)}")

    # Evaluation on the validation set
    print("Running Validation...")
    t0 = time.time()
    model_new.eval() # Put the model in evaluation mode

    eval_accuracy_new = 0
    nb_eval_steps_new = 0
    all_preds_new = []
    all_labels_new = []

    for batch in validation_dataloader_new:
        batch = tuple(t.to(device) for t in batch)
        b_input_ids, b_input_mask, b_labels = batch

        with torch.no_grad(): # Don't compute gradients
            outputs = model_new(b_input_ids,
                                token_type_ids=None,
                                attention_mask=b_input_mask)

        logits = outputs[0]
        logits = logits.detach().cpu().numpy()
        label_ids = b_labels.to('cpu').numpy()

        eval_accuracy_new += flat_accuracy(logits, label_ids)
        nb_eval_steps_new += 1

        preds = np.argmax(logits, axis=1).flatten()
        all_preds_new.extend(preds)
        all_labels_new.extend(label_ids)

    avg_val_accuracy_new = eval_accuracy_new / nb_eval_steps_new
    print(f"  Accuracy: {avg_val_accuracy_new:.2f}")
    print(f"  Validation took: {format_time(time.time() - t0)}")

    # Calculate Precision, Recall, F1-score
    precision_new, recall_new, f1_new, _ = precision_recall_fscore_support(all_labels_new, all_preds_new, average='binary', zero_division=0)
    print(f"  Precision: {precision_new:.2f}")
    print(f"  Recall: {recall_new:.2f}")
    print(f"  F1-score: {f1_new:.2f}")

print("\nTraining complete for the new model!")

# 6. Save the trained model to a *new* directory
output_dir_new = './new_model_save/'
if not os.path.exists(output_dir_new):
    os.makedirs(output_dir_new)

print(f"Saving new model to {output_dir_new}")
model_to_save_new = model_new.module if hasattr(model_new, 'module') else model_new
model_to_save_new.save_pretrained(output_dir_new)
tokenizer.save_pretrained(output_dir_new)

print("New model and tokenizer saved.")
```

```python
import torch
import numpy as np
from sklearn.metrics import accuracy_score, precision_recall_fscore_support
import time
import datetime

# Helper function for formatting time (re-defining for self-containment if not already in context)
def format_time(elapsed):
    elapsed_rounded = int(round((elapsed)))
    return str(datetime.timedelta(seconds=elapsed_rounded))

# Function to calculate accuracy (re-defining for self-containment if not already in context)
def flat_accuracy(preds, labels):
    pred_flat = np.argmax(preds, axis=1).flatten()
    labels_flat = labels.flatten()
    return np.sum(pred_flat == labels_flat) / len(labels_flat)

# Ensure the model is in evaluation mode
model_new_loaded.eval()

print("Evaluating new model on validation set...")
t0 = time.time()

# Initialize variables to store evaluation metrics
eval_accuracy_new = 0
nb_eval_steps_new = 0
all_preds_new = []
all_labels_new = []

# Iterate through the validation data
for batch in validation_dataloader_new:
    # Move batch to the appropriate device
    batch = tuple(t.to(device) for t in batch)
    b_input_ids, b_input_mask, b_labels = batch

    # Perform forward pass without computing gradients
    with torch.no_grad():
        outputs = model_new_loaded(
            b_input_ids,
            token_type_ids=None,
            attention_mask=b_input_mask
        )

    # Get logits and move to CPU
    logits = outputs[0]
    logits = logits.detach().cpu().numpy()
    label_ids = b_labels.to('cpu').numpy()

    # Calculate batch accuracy and accumulate
    eval_accuracy_new += flat_accuracy(logits, label_ids)
    nb_eval_steps_new += 1

    # Get predictions and store
    preds = np.argmax(logits, axis=1).flatten()
    all_preds_new.extend(preds)
    all_labels_new.extend(label_ids)

# Calculate average validation accuracy
avg_val_accuracy_new = eval_accuracy_new / nb_eval_steps_new
print(f"  Final Validation Accuracy (New Model): {avg_val_accuracy_new:.2f}")

# Calculate and print Precision, Recall, and F1-score
precision_new, recall_new, f1_new, _ = precision_recall_fscore_support(all_labels_new, all_preds_new, average='binary', zero_division=0)
print(f"  Final Validation Precision (New Model): {precision_new:.2f}")
print(f"  Final Validation Recall (New Model): {recall_new:.2f}")
print(f"  Final Validation F1-score (New Model): {f1_new:.2f}")

print(f"Evaluation of new model took: {format_time(time.time() - t0)}")
```
